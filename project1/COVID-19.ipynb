Find the continent has max human development index

```python
maxhdi=df.groupby('continent').agg(human_development_index=('human_development_index', 'max')).head(1)
print(maxhdi)
```

Find which continent has max gdp per capital

```python
maxgpc=df.groupby('continent').agg(gdp_per_capita=('gdp_per_capita', 'max')).head(1)
print(maxgpc)
```

Fiter the dataframe and update it

```python
df = df[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']]
df.head()
```

DATA CLEANING

Remove all dublicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

Find the missing values

```python
df.isnull().sum()
```

Remove all abservations where continent values are missing

```python
df.dropna(subset=['continent'])
```

Fill all missing value to 0

```python
df = df.fillna(0)
```

Convert date column to datetime format

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

Create new col month after extracting month date 

```python
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
df.head()
```

DATA AGGREGATION

Find max value in all col using groupby function on continent column

```python
df.groupby('continent').max().reset_index()
```

Store the new dataframe name df_gropuby

```python
df_groupby=df.groupby('continent').max().reset_index()
df_groupby
```

FEATURE ENGINEERING

Create new feature of ratio of total death to total cases 

```python
df_groupby['total_deaths_isto_total_cases'] = df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby['total_deaths_isto_total_cases']*100
```

```python
df_groupby
```

Oceania has the highest ratio of death to cases
